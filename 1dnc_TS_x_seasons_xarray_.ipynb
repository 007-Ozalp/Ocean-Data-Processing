```python
rcParams['figure.figsize'] = 18, 8
rcParams['lines.linewidth'] = 1
```

```python
#ds1_season = f['thetao'].groupby('time.season')['DJF'].mean('time')
ds1_season_inv = f['thetao'].groupby('time.season')['DJF']
```

```python
#ds1_season_inv.plot()
####################################################################################################
ds1_season_inv.plot(color='darkviolet', label='DJF')
####################################################################################################
title = "TS cmems reanalysis Temperature at Sea Surface from1987to2019 in AdriaticSea during Winter"
plt.title(title)
plt.legend();
```

```python
ds1_season_est = f['thetao'].groupby('time.season')['JJA']
```

```python
#ds1_season_est.plot()
####################################################################################################
ds1_season_est.plot(color='r',marker='o', label='JJA')
####################################################################################################
title = "TS cmems reanalysis Temperature at Sea Surface from1987to2019 in AdriaticSea during Summer"
plt.title(title)
plt.legend();
```

```python
ds1_season_prim = f['thetao'].groupby('time.season')['MAM']
```

```python
#ds1_season_prim.plot()
####################################################################################################
ds1_season_prim.plot(color='g',marker='o', label='MAM')
####################################################################################################
title = "TS cmems reanalysis Temperature at Sea Surface from1987to2019 in AdriaticSea during Spring"
plt.title(title)
plt.legend();
```

```python
ds1_season_aut = f['thetao'].groupby('time.season')['SON']
```
