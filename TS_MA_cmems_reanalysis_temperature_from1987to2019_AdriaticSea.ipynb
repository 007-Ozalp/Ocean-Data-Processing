```python
####################################################################################################
####################################################################################################
####################################################################################################
####################################################################################################


### Window Rolling Mean (Moving Average) through pandas

### Rolling mean:an average for a window of data

# TS in Original-black line
#TS with Moving Average period= MA60-dark violet line
#TS with Moving Average period= MA90-blue line
#TS with Moving Average period= MA120-dark yellow line
#TS with Moving Average period= MA180-lime line


####################################################################################################
####################################################################################################
####################################################################################################
####################################################################################################



import numpy as np
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
from matplotlib import rcParams
from cycler import cycler
from statsmodels.tsa.stattools import adfuller, kpss
```

```python
rcParams['figure.figsize'] = 18, 5
rcParams['lines.linewidth'] = 3
####################################################################################################
df = pd.read_csv('TS_cmems_reanalysis_temperature_from1987to2019_AdriaticSea.csv', index_col='DATE', parse_dates=True)
####################################################################################################
plt.title('Temperature at Sea Surface since 1987 over Adriatic Sea', size=20)
plt.plot(df, color='k', label='Original')
```

```python
### rolling mean:an average for a window of data
### Window Rolling Mean (Moving Average) 

df['MA60'] = df['TEMPERATURE'].rolling(window=60).mean()
df['MA90'] = df['TEMPERATURE'].rolling(window=90).mean()
df['MA120'] = df['TEMPERATURE'].rolling(window=120).mean()
df['MA180'] = df['TEMPERATURE'].rolling(window=180).mean()

```

```python
plt.title('TS cmems reanalysis temperature from1987to2019 AdriaticSea MA=60', size=20)
####################################################################################################
plt.plot(df['MA60'], color='darkviolet', label='MA60')
####################################################################################################
plt.legend();
```

```python
plt.title('TS cmems reanalysis temperature from1987to2019 AdriaticSea MA=90', size=20)
####################################################################################################
plt.plot(df['MA90'], color='b', label='MA90')
####################################################################################################
plt.legend();
```

```python
plt.title('TS cmems reanalysis temperature from1987to2019 AdriaticSea MA=120', size=20)
####################################################################################################
plt.plot(df['MA120'], color='y', label='MA120')
####################################################################################################
plt.legend();
```

```python
plt.title('TS cmems reanalysis temperature from1987to2019 AdriaticSea MA=180', size=20)
####################################################################################################
plt.plot(df['MA180'], color='lime', label='MA180')
####################################################################################################
plt.legend();
```
